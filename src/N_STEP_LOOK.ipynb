```python
from utils.step_helper import score_move
import numpy as np
import random 




def N_step_look(obs,config):
    valid_moves=[ col for col in range(config.columns) if obs.board[col] == 0 ]

    print(f"valid moves: {valid_moves}") 
    grid= np.asarray(obs.board).reshape(config.rows,config.columns)

    scores=dict(zip(valid_moves, [score_move(grid=grid,col=col,mark=obs.mark,config=config,nsteps=3) for col in valid_moves]))

    if scores:
        return random.choice(valid_moves)
    max_cols=[key for key in scores.keys() if scores[key] == max(scores.values())]

    chosen_move=random.choice(max_cols)

    

    return chosen_move
```
